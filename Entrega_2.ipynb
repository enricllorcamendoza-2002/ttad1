## 4.Combinando información en pandas

Dispones de los siguientes DataFrames:

```python
import pandas as pd

estudiantes= pd.DataFrame({
    "id":[1, 2, 3,4],
    "nombre":["Ana", "Luis","Marta", "Jorge"]
 })

notas =pd.DataFrame({
    "id":[1, 2, 2,5],
    "asignatura":["Matemáticas","Historia","Química","Biología"],
    "nota":[8.5, 7.0,6.5, 9.0]
 })
```

```python
estudiantes
```

```python
notas
```

(a) Realiza un merge interno (inner) entre ambos DataFrames. ¿Cuántas filas se obtienen? ¿Por qué?

```python
estudiantes.merge(notas, left_on='id', right_on='id', how='inner')
```

Observamos que solamente se obtienen 3 filas. Esto se debe a que solamente hay 3 filas donde coincida la variable id. En el caso de estudiantes tenemos 2 individuos con id 1 y 2 y en notas 2 con observaciones con id 2 y una con id 1. Las id 1 y 2 son las únicas que coinciden. id 3 y 4 no se encuentran en notas y id 5 no se halla en estudiantes.

(b) Realiza un merge externo (outer). Explica las filas que aparecen con valores nulos.

```python
estudiantes.merge(notas, left_on='id', right_on='id', how='outer')
```

Dado que, como hemos explicado en el apartado anterior, hay datos de id que no están en uno u otro fichero, los datos para esos ficheros que no encuentran un "match" en las columnas del otro DataFrame quedan como valores desconocidos NaN cuando hacemos un Outer Merge, que muestra todos los datos de los dos conjuntos, hagan o no "match" el uno con el otro.

(c) Realiza un left merge usando estudiantes como tabla izquierda. ¿Qué información se pierde respecto al outer?

```python
estudiantes.merge(notas, left_on='id', right_on='id', how='left')
```

El Left Merge coge como DataFrame base estudiantes (izquierda) y a este le añade las columnas del DataFrame notas (derecha) que hagan "match", por lo que, como el id 5 no aparece en estudiantes, se pierde la última fila del Outer Merge

(d) Realiza un right merge usando notas como tabla derecha. ¿Qué estudiantes desaparecen en el resultado?

```python
estudiantes.merge(notas, left_on='id', right_on='id', how='right')
```

Ahora es al contrario, el Right Merge, hace que tan solo se complete la información para los id que ya tenemos de la tabla derecha, notas, por lo que en este caso se pieden los datos cuyos id eran 3 y 4. En definitiva, se pierden los datos del id y nombres de Marta y Jorge (2 filas).

(e) Añade una columna llamada curso al DataFrame estudiantes. Después, combina los datos utilizando un merge mútiple por id y explica el resultado.

```python
#1) Vamos a añadir la nueva columna con datos inventados

estudiantes2 = estudiantes.assign(curso=[4,5,6,2])
estudiantes2 #DataFrame Modificado
```

```python
#2) Merge Múltiple por id y explicar resultados: inner merge es el merge por defecto
estudiantes2.merge(notas, left_on='id', right_on='id')
```

De nuevo, los resultados del 'inner' merge resultan en 3 filas ya que tan solo los id 1 y 2 coinciden en ambos DataFrames, con 2 personas con dos datos para la persona con id 2 (Luis) en el DataFrame notas.

## 5.Agregaciones y operaciones con groupby

En este ejercicio se trabajará con un conjunto de datos que contiene información sobre ventas realizadas por una empresa en distintas regiones. Considera el siguiente DataFrame:

```python
import pandas as pd

ventas= pd.DataFrame({
    "region":["Norte","Norte","Sur","Este","Sur","Este","Oeste","Oeste","Norte","Sur"],
    "vendedor":["Ana","Luis","Ana","Jorge","Luis","Ana","Marta","Luis","Ana","Jorge"],
    "ventas":[1200, 800,950, 1100,400, 1600,700, 450,895, 900],
    "producto":["A","A","B","A","B","A","B","A","B","B"]
})
ventas
```

Realiza las siguientes tareas:
 (a) Obtén el total de ventas por región.


```python
#Agrupamos por region y calculamos la suma de ventas 
gr = ventas.groupby(['region'])["ventas"].sum()
gr
```

 (b) Calcula la media, el valor máximo y el mínimo de las ventas agrupadas por vendedor.


```python
#Agrupamos por vendedor y mostramos la medio, el máximo y mínimo de las ventas
gr = ventas.groupby("vendedor")["ventas"].agg(["mean", "max", "min"])
gr
```

 (c) Calcula el número de ventas realizadas por cada combinación de region y producto.


Nota: Si interpretamos que "ventas" es el número de ventas que se realizan de cada producto por cada vendedor habría que usar sum()

```python
#Agrupamos por region, luego producto y damos el total de ventas
gr = ventas.groupby(["region", "producto"])["ventas"].sum()
gr
```

Pero, si interpretamos que "ventas" es el importe de cada venta, debemos usar count() para obtener el número de ventas realizadas

```python
#Agrupamos por region, luego producto y damos el total de ventas
gr = ventas.groupby(["region", "producto"])["ventas"].count()
gr
```

 (d) Crea una función llamada rango_ventas en un fichero.py que calcule la diferencia entre el valor máximo y mínimo de un grupo. Aplícala a las ventas agrupando por región.


```python
#Llamar fichero .py
import p2_funciones as p2

#Aplicar la función sobre las ventas, que están agrupadas por región.
rango_por_region = ventas.groupby("region")["ventas"].apply(p2.rango_ventas)
rango_por_region
```

 (e) Identifica la región con el mayor número total de ventas y muestra solo esa región.

```python
#(1) Agrupar por región con el total de ventas
gr = ventas.groupby("region")["ventas"].sum()
print(gr)

print("-"*10)

#(2) Ordenar de más a menos ventas
gr_sorted = gr.sort_values(ascending= False)
print(gr_sorted)

#(3) Hallar el maximo
gr.max()

print("-"*10)

#(4) Mostrar solo esa región
print(gr[gr==gr.max()])

print("-"*10)

#(5) Región donde máximo (nombre de la región)
print("Región con Máximo número de ventas: ", gr_sorted.index[0])

#Alternativa...

print("Región con Máximo número de ventas: ", gr[gr==gr.max()].index[0])



```
