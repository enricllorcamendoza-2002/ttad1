Tenemos los estadísticos descriptivos de cada columna. Count mide el número de observaciones por columna (número de filas), Mean es la media de los valores (el promedio) y std la desviación estándar. Min y Max miden el máximo y el mínimo valor respectivamente. 25%, 50%, 75% representan percentiles, es decir, muestran los valores los cuales tienen a su izquierda, es decir como valores menores, el 25, 50 y 70 porciento del resto de valores.

 (b)Convierte la columna quality en una columna categ´orica con tres categor´ıas:

 • bueno
 
 • regular
 
 • malo

```python
#Solucion

df["quality"] = pd.cut(df.quality, 3, labels = ["malo", "regular", "bueno"])

#De todas las filas, estas columnas
df.loc[:, ["quality"]].head(5)
```

(c) Selecciona los vinos clasificados como malo y elim´ınelos del DataFrame.

```python
#Seleccionar filas con vinos malos

fila_borrar = df[df["quality"]=="malo"]
fila_borrar.index
```

```python
#Borrar filas donde calidad es mala

df_no_malos = df.drop(fila_borrar.index)
df_no_malos
```
