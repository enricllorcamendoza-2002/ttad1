 ## 1. Creaci´on de DataFrames

```python
import pandas as pd
import numpy as np
```

(a) Crea un DataFrame df1 con una columna llamada numeros que contenga los
 valores del 0 al 19

```python
df1 = pd.DataFrame(
    {"Col1":np.arange(20)})

df1.tail(3)
```

(b) Crea un DataFrame df2 con una columna aleatorios de tama˜no 10 con
 n´umeros aleatorios entre 0 y 1

```python
np.random.seed(1)

df2 = pd.DataFrame(
    {"Col1":np.random.random(10)})

df2
```

 (c) Combina los DataFrames anteriores df1 y df2 en un nuevo DataFrame llamado
 df3.

```python
df3 = pd.concat([df1,df2]) #Los junta en la misma columna ya que en ambos era "Col1"
df3 = df3.reset_index()
df3
```

 (d) Guarda el DataFrame df3 en un fichero .csv.

```python
df3.to_csv('data/df3_file.csv')
```
