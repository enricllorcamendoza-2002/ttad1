#### D. ¿Cuál es el rango de años en la antigüedad del vehículo? ¿Cuál es el año más frecuente?


Vehicle Year = año de fabricación/matriculación del vehículo.

```python
#0) Análisis preliminar

#Examinemos los años de amtigüedad presentes en la muestra (ordenados)
df["Vehicle Year"].sort_values().unique() #Suponer que los valores con 0 no hay que tenerlos en cuenta
```

```python
print(df["Vehicle Year"].sort_values().value_counts().head()) #Año cero con abrumadora mayoría, parece claro que estos casos son los casos donde el año de antigüedad del vehículo es desconocido.
print("_"*20)
print(df["Vehicle Year"].sort_values().value_counts().tail())
```

Vistos estos resultados parece razonable pensar que el año cero se ha puesto para los registros donde el año es desconocido. Por ello, tomaremos el valor 0 como si fuese un valor desconocido o NaN y eliminaremos las observaciones con dicho valor para este análisis por lo que el rango de años será entre 2069 y el año 1980.

```python
#1) Ordenar por año de antigüedad del vehículo
df_ord = df.sort_values(by="Vehicle Year", ascending=True)
df_ord

#2) Eliminar los valores donde el año es cero: Supongo que esos años deberían ser NaN o son valores desconocidos
df_ord.drop(df_ord[df_ord["Vehicle Year"] == 0].index, inplace=True)
df_ord

#RANGO AÑOS ANTIGÜEDAD
#3) Restar "último año" del "primero"
rango = df_ord["Vehicle Year"].values[-1]-df_ord["Vehicle Year"].values[0]
print("Primer Año:", df_ord["Vehicle Year"].values[0])
print("Ultimo Año:", df_ord["Vehicle Year"].values[-1])
print("Rango de años conocido: ", rango)

#AÑO MÁS FRECUENTE
#4) Value counts de df["Vehicle Year"]
mas_frec = df_ord["Vehicle Year"].value_counts()
print("Año más frecuente conocido: ", mas_frec.index[0])
```

Alternativamente, si lo midiesemos por Issue Date y no Vehicle Year

```python
#NOTA: Si se tratase de Issue Date y no de Vehicle Year...

# Pasar indice a fecha
from pandas import DatetimeIndex

#1) Convertir date en Indixe que diferenica año, mes y día -> Detecta el formato PERO NECESITO SUFICIENTES MUESTRAS
df.index = DatetimeIndex(df["Issue Date"])
df.head()

#RANGO AÑOS ANTIGÜEDAD

#1) Ordenar por años de anitgüedad -> Ordenar filas por el índice (de menor a mayor)
df_ordenado = df.sort_index()


#2) Restar última fecha de primera fecha: Rango años antigüedad
difer =df_ordenado.index.year[-1]-df_ordenado.index.year[0]
print("Rango de años :", difer)

#AÑO MÁS FREUCENTE

#1) Value counts de index.year
mas_frec = df.index.year.value_counts()
print("Año más frecuente: ", mas_frec.index[0])
```

#### J. ¿Hay vehículos que han recibido más de una multa ?


```python
#ANALISIS PREELIMINAR: Ver cuantas multas hay por vehículo
df["Plate ID"].value_counts()
```

Cada Vehículo Individual tiene un Plate ID especifico de ese coche. En este análisis eliminaremos los BLANKPLATES y los tomaremos como valores desconocidos ya que pueden ser, o no, el mismo coche por lo que trabajamos con información desconcida.

```python
#0) Eliminar BLANKPLATES ya que pueden ser coches diferentes
df_clean = df[df["Plate ID"] != "BLANKPLATE"].copy()

#1) El nº de Multas por vehículo
multas_por_vehiculo = df_clean["Plate ID"].value_counts()

#Alternativa: df_clean.groupby(["Plate ID"])["Plate ID"].count().sort_values(ascending=False)

#2) ¿Hay Vehículos que han recibido más de una multa?
cuantos = multas_por_vehiculo[multas_por_vehiculo > 1].count() #Cuenta Filas donde se cumple

#Alternativa: (multas_por_vehiculo>1).sum() // True=1, False=0

print("Hay", cuantos, "vehículos como mínimo que han recibido más de una multa.")

#3) Se cumple que hay algunos que han recibido más de una multa?
print("Hay vehículos con más de una multa que sepamos?", cuantos>1)

#Alternativa: (multas_por_vehiculo > 1).any()
```

####  L. Transmite una idea o mensaje con una visualización. Diseña y comenta mensaje y posteriormente, genera la visualización.


```python
#Cargar Fichero
import pandas as pd
import numpy as np

df = pd.read_csv("sample_NYC_parking_issues.csv", sep=";") #La separación queda hecha por ";"
df.head()  # primeras filas para comprobar que se ha cargado bien
```

#### FIGURA 1: Número de Multas por Marca (Vehicle Make) de las 10 más frecuentes

La Figura 1 muestra el número de multas de aparcamiento por marca de vehículo para las 10 marcas más frecuentes. Se observa que unas pocas marcas concentran una proporción elevada de las sanciones, lo que sugiere que la distribución de multas no es uniforme entre fabricantes. Esta visualización permite identificar qué tipos de vehículos aparecen con más frecuencia en el registro de infracciones y plantea la pregunta de si esto se debe a su mayor presencia en las calles o a patrones de uso y comportamiento diferentes de los clientes de dicha marca.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#1) Encontrar 10 marcas más populares
top10 = df["Vehicle Make"].value_counts().head(10)
print(top10)

#2) Representar 
ax = top10.plot(kind ="bar")
ax.set_ylabel("Número de Multas")

plt.title("Barras: Nº de Multas por Marca de las 10 más frecuentes.")

plt.show()
```

### FIGURA 2: Distribución código de Violación (Violation Code) y año del vehículo (Vehicle Year) para los 3 colores más frecuentes de coches, mostrando dichos colores.

Esta figura muestra cómo se distribuyen las infracciones según la antigüedad del vehículo, lo que permite comparar los patrones de coches más nuevos frente a los más antiguos e identificar qué rangos de años concentran más multas. La codificación por color añade una tercera dimensión: podemos ver de qué color tienden a ser los vehículos sancionados en cada rango de antigüedad y si ciertos colores aparecen asociados a determinados tipos de infracción.


Para evitar resultados poco realistas, he eliminado los registros con Vehicle Year = 0, que interpreto como un valor desconocido del año del vehículo

```python
df["Vehicle Year"].unique()
```

```python
#PASO 1: Quedarnos solamente con los 3 colores más frecuentes

#1) Top 3 de colores
top3 = df["Vehicle Color"].value_counts().head(3) 
print(top3)

#2) Quedarme solamente con el top 3 de colores
df_top = df[df["Vehicle Color"].isin(top3.index)]

#Comprobamos
df_top["Vehicle Color"].unique()
```

```python
# https://seaborn.pydata.org/generated/seaborn.rugplot.html
import seaborn as sns; sns.set_theme()

# PASO 2: GRAFICO PUNTOS / SCATTERPLOT

#1) Definir forma y contenido
fig, ax = plt.subplots(figsize=(10, 8)) #tamaño figuras

#2) Hacer scatter plot de Ciudad y Puntos 
sns.scatterplot(data=df_top, x="Violation Code", y="Vehicle Year", hue="Vehicle Color")

#3) Añadir rugplot a la izquierda (que apila los puntos)
sns.rugplot(data=df_top, x="Violation Code", y="Vehicle Year", hue="Vehicle Color") #rugplot lo del lado izquierdo (apila los puntos)

#4) Añadir labels
plt.title("Distribución de Año del Vehículo por Codigo de Infracción")
plt.xlabel("Violation Code")
plt.ylabel("Vehicle Year")

plt.show()
```

Todavía tenemos el problema del año cero que seguramente son valores desconocidos por lo que eliminaremos dichas observaciones para facilitar la visualización

```python
#PASO 3: Eliminar Observaciones con año cero

# Eliminar los valores donde el año es cero: Supongo que esos años deberían ser NaN o son valores desconocidos
df_top.drop(df_top[df_top["Vehicle Year"] == 0].index, inplace=True)

# VOLVEMOS GRAFICO PUNTOS / SCATTERPLOT

#1) Definir forma y contenido
fig, ax = plt.subplots(figsize=(10, 8)) #tamaño figuras

#2) Hacer scatter plot de Ciudad y Puntos 
sns.scatterplot(data=df_top, x="Violation Code", y="Vehicle Year", hue="Vehicle Color")

#3) Añadir rugplot a la izquierda (que apila los puntos)
sns.rugplot(data=df_top, x="Violation Code", y="Vehicle Year", hue="Vehicle Color") #rugplot lo del lado izquierdo (apila los puntos)

#4) Añadir labels
plt.title("Distribución de Año del Vehículo por Codigo de Infracción")
plt.xlabel("Violation Code")
plt.ylabel("Vehicle Year")

plt.show()
```

Vemos que los vehículos más nuevos (de 1995 hasta 2014), concentran la mayoría de las multas. También observamos una concentración de ciertos tipos de infracciones especificas con 3 grupos marcados: las infracciones de la 0 a la 20, de la 35 a la 55 y de la 65 a la 85 aproximadamente, son grupos especificos de infracciones bastante más frecuentes que el resto apra los 3 colores de vehíuclos más frecuentes. También podemos ver que los coches más nuevos tienden a tener el color WG y las más antiguos más al color WHITE.
