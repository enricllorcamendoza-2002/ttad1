2) Indexación y slicing

```python
x = np.arange(10, 30)
print(x)

#(a) Muestra los elementos del índice 3 al 8
print(x[3:9]) #indicamos hasta nueve porque muestra hasta el índice anterior a 8 inclusive

print("-"*10)

#(b) Muestra los elementos pares del array.

#Para ser pares modul: tiene que ser 0
modul = x%2
print(modul)

#Muestra los que son pares
print(x[modul==0])

print("-"*10)

#(c) Cambia el valor del último elemento a 100.

#Longitud de x y Ultima posición
last_pos = len(x)-1

#Modificar último valor
x[last_pos]=100
print(x)
```

```python
#Supón que tienes un array que representa las temperaturas diarias (°C) de una ciudad durante una semana:
temps = np.array([18, 20, 22, 21, 19, 17, 16])

#(a) Calcula la temperatura media semanal.
media = np.mean(temps)
print(media)
print("-"*10)

#(b) Muestra qué días estuvieron por encima del promedio.
print(temps[temps>media])
print("-"*10)

#(c) Convierte todas las temperaturas a °F usando la fórmula:
Faren = temps*(9/5) + 32
print(Faren)
print("-"*10)

#(d) Agrega una nueva temperatura (día 8 = 23°C) al array original.

#(1) Saber longitud del array
longit = len(temps)
print(longit)

#(2) Añadir al final 
np.append(temps, 23)    # Usamos append
```

```python
import numpy as np
```

```python
data1 = np.loadtxt("data/datos_1.csv", delimiter=",", usecols=(1,))
print(data1)
print(data1.shape)
print("-"*10)

data2 = np.loadtxt("data/datos_2.csv", delimiter=",", usecols=(1,), skiprows=1, encoding="latin-1")
print(data2)
print(data2.shape)
```

```python
import numpy as np

with open("data/datos_1.csv", encoding="latin-1") as f:
    rows = [line.strip().strip('"').split(',') for line in f if line.strip()]

data1 = np.array(rows, dtype=float)  # matriz 2D
print(data1)

with open("data/datos_2.csv", encoding="latin-1") as f:
    rows = [line.strip().strip('"').split(',') for line in f if line.strip()]

data2 = np.array(rows, dtype=float)  # matriz 2D
print(data2)
```

```python
#(a) El conjunto de variables estadísticas básicas:

# Media.
print("Media de Datos 1: ", np.mean(data1))
print("Media de Datos 2: ", np.mean(data2))
print("-"*10)

# Mediana.
print("Mediana de Datos 1: ", np.median(data1))
print("Mediana de Datos 2: ", np.median(data2))
print("-"*10)

# Valor máximo y mínimo.
print("Max de Datos 1: ", np.max(data1))
print("Max de Datos 2: ", np.max(data2))
print("Min de Datos 1: ", np.min(data1))
print("Min de Datos 2: ", np.min(data2))
print("-"*10)

# Número de elementos presentes en cada fichero.
print("Numero de elementos en Datos 1: ", data1.size)
print("Numero de elementos en Datos 2: ", data2.size)
```

```python
#Añadir filas
data3 = np.concatenate((data1,data2), axis=0)
data3

# Media.
print("Media de Datos 3: ", np.mean(data3))
print("-"*10)

# Mediana.
print("Mediana de Datos 3: ", np.median(data3))
print("-"*10)

# Valor máximo y mínimo.
print("Max de Datos 3: ", np.max(data3))
print("Min de Datos 3: ", np.min(data3))
print("-"*10)

# Número de elementos presentes en cada fichero.
print("Numero de elementos en Datos 3: ", data3.size)
```

```python
#(b) La intersección de los elementos de la primera fila de los dos arrays.
print(np.intersect1d(data1[0], data2[0]))

#(c) La unión de los elementos de las primeras filas de los dos arrays (sin duplicados).
print(np.union1d(data1[0], data2[0]))

```

```python
#ALTERNATIVA 1
import numpy as np
a = np.ascontiguousarray(data1)
b = np.ascontiguousarray(data2)

dt = np.dtype([('f'+str(i), a.dtype) for i in range(a.shape[1])])

rows = np.intersect1d(a.view(dt).ravel(), b.view(dt).ravel())
rows = rows.view(a.dtype).reshape(-1, a.shape[1])
rows
```

```python
#ALTERNATIVA 2
a, b = map(np.ascontiguousarray, (data1, data2))          # 2D, mismo nº de columnas
dt = np.dtype((np.void, a.dtype.itemsize * a.shape[1]))   # “fila” como bloque de bytes
rows = np.intersect1d(a.view(dt).ravel(), b.view(dt).ravel()) \
           .view(a.dtype).reshape(-1, a.shape[1])

print(rows)
```

```python
#(d) Filas repetidas en ambos ficheros (de uno al otro)
rows = np.intersect1d(data1, data2, axis=0)
print(rows)
```

```python
#(d) Filas repetidas en ambos ficheros (entiendo que se repiten, ya sea en uno como en otro)

#(1) Filas unicas y su frecuencia
unique_3, freq_3 = np.unique(data3, axis=0, return_counts=True)

#(2) Filas Unicas
print(unique_3)

#(3) Veces que se repite cada una
print(freq_3)
print("-"*10)

#(4) Mostrar filas que cumplen esas condiciones: Filas de posiciones donde la frecuencia es mayor a 1
print(unique_3[np.argwhere(freq_3>1)])
```

```python
# 1. Convierte las filas de cada array a un conjunto de tuplas

# Esto permite que Python trate cada fila como un elemento único e inmutable
set1 = set(map(tuple, data1))
set2 = set(map(tuple, data2))

# 2. Encuentra la intersección de los dos conjuntos

# El resultado serán las filas comunes (como tuplas)
interseccion_tuplas = set1.intersection(set2)


# 3. Convierte las tuplas resultantes de nuevo a un array de NumPy
filas_comunes = np.array(list(interseccion_tuplas))

print("Filas Comunes (Intersección):")
print(filas_comunes)
```
